```python
import pandas as pd
import json
```

```python
df = pd.read_excel('sheet.xlsx')
```

```python
df
```

```python
df["URL_original"].fillna(df["URL_sinequa"], inplace=True)
```

```python
new = df[['Name of Resource', 'Sinequa Config Guess', 'URL_original', 'turned_on', 'Division']]
```

```python
new.rename(columns={'Name of Resource': 'name',
'Sinequa Config Guess': 'config_folder',
'URL_original': 'url',
'turned_on': 'turned_on',
'Division': 'division',}, inplace=True)
```

```python
new
```

```python
new.turned_on = new.turned_on.fillna(False)
```

```python
new.turned_on = new.turned_on.replace(1.0, True)
new.turned_on = new.turned_on.replace(0.0, False)
new.turned_on.value_counts()
```

```python
new.division = new.division.replace('Planetary', 'Planetary Science')
```

```python
new.url = new.url.fillna('')
```

```python
new.config_folder = new.config_folder.fillna('')
```

```python
new
```

```python
new.url = new.url.replace('Use custom connector built by team', '')
```

```python
new
```

```python
# remove multi urls
mask = new.url.apply(lambda row: row.startswith('['))
column_name = 'url'
new.loc[mask, column_name] = ''
```

```python
new.to_csv('ready_to_load.csv', index=False)
```

```python
def create_fixture(row):
    division_mapper = {"Planetary Science": 5,
    "Earth Science": 4,
    "Biological and Physical Sciences": 3,
    "Heliophysics": 2,
    "Astrophysics": 1,}
    return {
            "model": "sde_collections.collection",
            "fields": {
                "name": row["name"],
                "config_folder": row["config_folder"],
                "url": row["url"],
                "division": division_mapper[row["division"]],
                "turned_on": row["turned_on"]
            }
    }
```

```python
new['json'] = new.apply(lambda row: create_fixture(row), axis=1)
```

```python
with open('../sde_collections/fixtures/collections.json', 'w') as fixturefile:
    fixturefile.write(json.dumps(new.json.to_list()))
```
