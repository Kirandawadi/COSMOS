```python
url_field = 'url1'
collection_config_folders = ['gcn_circulars']
for collection_congif_folder in collection_config_folders:
    page = 1
    url_info_dict = {}
    while True:
        response = api.query(page=page, collection_config_folder=collection_congif_folder)
        if response.get("cursorRowCount", 0) == 0:  # Safeguard against missing 'cursorRowCount'
            break;
        for record in response.get('records', []):  # Safeguard against missing 'records'
            url = record.get(url_field)
            title = record.get('title')
            if url and title:  # Ensure both url and title are present
                url_info_dict[url] = {'title': title}
        print(f'Finished fetching page : {page}')
        page += 1
    print(f'Number of urls is : {len(url_info_dict)}')
    break
```
