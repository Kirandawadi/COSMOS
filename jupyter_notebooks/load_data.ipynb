```python
import pandas as pd
import xmltodict
```

```python
df = pd.read_excel('nasa_power.xlsx')
```

```python
new = df[df['Index?']=='no'][['Full Url', 'Index?', 'How to Exclude']].fillna('exact')
```

```python
new
```

```python
exclude_patterns = []
for row in new.iterrows():
    if row[1][2] == 'exact':
        exclude_patterns.append(f'{row[1][0]}*')
    else:
        exclude_patterns.append(row[1][2])
```

```python
exclude_patterns = list(set(exclude_patterns))
```

```python
"""<UrlLinkExcluded>a*</UrlLinkExcluded>
<UrlLinkExcluded>/docs/*</UrlLinkExcluded>
<UrlLinkExcluded>https://somewebsite.com/something.html*</UrlLinkExcluded>"""
```

```python
with open('../sinequa_configs/sources/SMD/NASA_Power/default.xml') as config:
    conf = config.read()
```

```python
conf_dict = xmltodict.parse(conf)
```

```python
for pattern in exclude_patterns:
    conf_dict['Sinequa']
```

```python
conf_dict['Sinequa']['UrlLinkExcluded'] = exclude_patterns
```

```python
conf_dict['Sinequa']['UrlLinkExcluded']
```

```python
excluded_extensions = [
    '*.js',
    '*.css',
    '*.jy',
    '*.ico',
    '*.gz',
    '*.xml'
]
```

```python
conf_dict['Sinequa']['UrlLinkExcluded'].extend(excluded_extensions)
```

```python
print(xmltodict.unparse(conf_dict, pretty=True))
```
